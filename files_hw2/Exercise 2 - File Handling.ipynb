```python
# 10. Combine two CSVs, sales_jan.csv and sales_feb.csv, into a new CSV named sales_janfeb.csv.
df1 = pd.read_csv('./sales_jan.csv')
df2 = pd.read_csv('./sales_feb.csv')

df = pd.concat([df1, df2])
df.to_csv('sales_janfeb.csv', index=False)
```

```python
# 11. Write a script that reads the sales.csv file. After loading it filter out records where the quantity is more than 50. 
# Print the result of the filtered data.
df = pd.read_csv('./sales.csv')

quantity = df[df['quantity'] > 50]
print(quantity)
```
